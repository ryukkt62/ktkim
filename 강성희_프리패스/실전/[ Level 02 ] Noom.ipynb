### 서비스가 굉장히 크고 트래픽이 많으면, 하나의 컴퓨터로 처리할 수 없음
### 그러므로 분산처리함, 여러 대의 컴퓨터에서 데이터를 분산시켜서 저장하고 가져옴
### 즉 데이터를 제공해 줄 때 묶어서 줄 수도 있지만 떨어뜨려서 줄 수도 있음
### 또 데이터를 제공해 줄 때 큰 거 하나를 제공해주는 것보다 작은 거 여러 개를 주는 게 더 편함

```python
# concat == concatenate

order = pd.concat([order201701, order201702])
order
```

```python
# 이 정보는 중요한 정보는 아니기 때문에 따로 빼뒀다가 필요할 때 읽어와서 합칠 수 있음

url = 'https://bit.ly/dsa-02-order-how'

how = pd.read_csv(url, index_col = 'id')
how
```

```python
# 오른쪽에다 합침, columns 대신에 숫자 1을 집어넣어도 됌

pd.concat([order, how], axis = 'columns') # default는 axis = 'index' 혹은 숫자 0
```

### 합칠 때 index를 명시적으로 지정해주지 않았다면 순서대로 합치고 빈 값은 NaN으로 만듬
### 지금처럼 index를 id로 지정해주면 id가 매칭되는 순으로 합쳐줌

## 합칠 때 중요한 것
## 1. 공통되는 것을 찾아야 함 
### 여기선 user_id가 user 데이터에서의 id(index로 쓰이는)와 매칭됌
### 또한 product_id가 product 데이터의 id(index로 쓰이는)와 매칭됌
### 보통 이러한 사항은 데이터를 제공해 주는 사람이 설명해주는 경우가 많음

```python
# order와 user를 합칠 때

pd.merge(order, user, left_on = 'user_id', right_on = 'id') 
# order가 먼저 들어갔으니 왼쪽, user가 오른쪽
# 즉 order의 'user_id'와 user의 'id'가 서로 매칭됌
```

```python
# 합칠 때 key가 다르지 않고 같으면 그냥 on으로 쓸 수 있음

user['user_id'] = user.index

user
```

```python
pd.merge(order, user, on = 'user_id') 
```

```python
order
```

```python
user
```

```python
# order와 user 사이에 'user_id' 말고는 서로 이름상으로 겹치는 게 없음
# 이럴 때는 자동으로 key를 찾아주기도 함(key를 user_id로 해서)
# 공통 column이 아닌데 우연찮게 이름만 같은 경우도 있을 수 있으므로 주의
# 예를 들면 order의 'price'와 product의 'price'

pd.merge(order, user)
```

```python
# 또 한가지 주의할 점
# key를 제대로 세팅안하면 뜬금없이 price를 합칠 수 있음

pd.merge(order, product)
```

```python
# key가 아닌데 이름이 같은 경우('price'), 이럴 경우 왼쪽은 x 오른쪽은 y가 들어감
# 이런 경우 column의 이름을 바꿔주거나 그래야 함

pd.merge(order, product, left_on = 'product_id', right_on = 'id')
```

```python
pd.merge(order, user, on = 'user_id')
```

### 위의 order데이터에서 user_id가 8인 사람(주소: 대구, 결제가격: 650)인 사람이 합친데이터에서는 없음
### order와 user가 서로 동시에 존재하는 데이터도 있겠지만 없는 데이터도 있음
### user데이터에서는 Moon이란 사람이 있는데 합친 데이터에는 없음
### 이런 경우 합치는 전략이 달라짐

```python
# inner는 교집함, 양쪽 다 있어야 함
# 둘 중 하나라도 없으면 버림

pd.merge(order, user, on = 'user_id', how = 'inner') 
```

```python
# outer는 합집함, 없는 경우 NaN으로 표기

pd.merge(order, user, on = 'user_id', how = 'outer') 
```

```python
# 'left'는 왼쪽은 살리고 오른 쪽은 버림

pd.merge(order, user, on = 'user_id', how = 'left') 
```

```python
# 'right'는 오른쪽은 살리고 왼쪽은 버림

pd.merge(order, user, on = 'user_id', how = 'right') 
```

# Tip!
### 분석의 main이 되는 주제에 해당하는 데이터를 왼쪽에 넣고 left merge 하는 걸 추천
### 이렇게 하면 대부분 맞음
### 이게 안될 경우 right이나 inner, outer 사용
### 보통 inner를 많이 쓰게 되고 그 다음 right을 많이 씀. outer는 거의 쓰이지 않음
