* 데이터 사이언스 세계에서 100%는 존재하지 않음. 99%는 존재할 수 있음(근사하는 개념이기 때문에)
* 그렇기 때문에 100% 나온 경우 2가지로 생각할 수 있음
* 1. PassengerID 하나하나 고치고 제출하고 고치고 제출하고 약 400번 반복(올바른 데이터 분석 방법이 아님. 일종의 cheating) 이걸 막기 위해 하루에 제출할 수 있는 횟수는 10번으로 제한

* 피봇테이블에서  행을 하나 추가했는 데 에러가 날 수 있음(예시: #DIV/0!) 이런 경우 공식 넣어준 부분을 삭제해주면 됨
* %가 소수점 엄청 많이 나오는 경우(예시: 73.0000000%) 이런 경우 위에 123버튼 클릭 자동에 체크

## 두번째 가설: 어린 아일수록 생존 가능성이 높을 것이다

* 어떤 가설이 하나 맞아떨어지면 이것과 연관된 다른 가설들을 연쇄적으로 실험함. 그러면 계속 맞아떨어질 확률 높아짐. 그러므로 두번째 가설은 어린 아일수록 생존 가능성이 높을 것이다.
* 아까 그 Sex 피봇테이블 편집기에서 Sex에 이어 Age도 행에다가 추가
* 여기서 문제가 생김. 너무 잘게 쪼개짐. 제대로 볼 수가 없음 ㅠ
* 숫자 데이터(Age, Fare 등)를 피봇 테이블에 넣으면 너무 잘게 쪼개지는 문제가 있음
* 이런 경우 train 데이터에 새로운 컬럼을 추가하고 구역을 나눠줘야 함. 그리고 이 구역을 기준으로 피봇 테이블을 만들어 줘야함
* 이전에 쓰던 피봇 테이블은 지우고 train 데이터에서 Age 옆에 새로운 컬럼 생성(마우스 오른쪽 왼쪽에 1열 삽입, Child), 오른쪽에 삽입하면 귀찮은 일 발생. 나중에 설명
* 새로 생성된 Child 컬럼에 =IF(G2 < 18, 1, 0) 18세 이상이면 1 이하면 0
* Age가 비어있는 칸은 Child가 1로 들어가는 데 이건 나중에 해결
* 십자가 더블클릭으로 전체 열에 넣음. 이때 아까 Child 컬럼을 Age오른쪽에 넣었을 경우 더블클릭을 해도 전체 생성이 안되고 Age가 비어있는 칸 위까지 밖에 생성이 안됨. 그 이유: 십자 더블클릭의 룰 - 왼쪽 컬럼의 비어있는 칸 위까지만 채움
* 아까 처럼 A와 1번 사이의 빈칸 클릭해서 전체 선택 - 데이터에서 피봇 테이블
* 값에는 Survived 2개, 위에는 SUM, 아래는 COUNT
* 분석하고 싶은 걸 행에 넣음. Sex와 Child를 넣음
* 생존률 구하는 공식 삽입 E4에 = C4 / D4
* 가설 검증: 여성 어린아이(68%)가 여성 어른(77%)보다 생존률이 더 낮음 -> 가설 검증 실패

* 가설은 틀렸지만 많은 걸 생각해 볼 수 있음.
* 1. 나이가 생존자와 사망자를 구분하는 데 중요하지 않을 수 있다.

* 3. 나이는 중요한 정보가 아닐 수 있지만 나이와 다른 정보를 조합하면 중요한 정보가 나올 수도 있다.

## 세번째 가설: 객실 등급이 좋을수록 탈출이 수월할 것이다(구조 장비 등)

* 피봇 테이블 편집기에서 Child를 빼고 Pclass 추가.(Sex와 Pclass 2개로 진행)
* 여성의 경우 1등급일 경우 생존률이 96%, 3등급일 경우 50%, 남성도 1등급이 훨씬 높음
* 가설 검증
* 첫번째 예측: 다 죽는다
* 두번째 예측: 여성이면 살고 남성이면 죽는다
* 세번째 예측: 여성이면 살고 남성이면 죽는다. 단, 여성이여도 3등급이면 죽는다
* 여성이여도 3등급이면 죽는 조건 실행
* Test 파일에서 원본을 복사하여 새 시트 만듬(Sex + Pclass)
* Pclass 왼쪽에 1열 삽입 이름: Survived
* 엑셀 공식을 넣는데, 엑셀 공식이 이제 좀 복잡함
* 이 때 팁: 긴 엑셀 공식을 한번에 짜고 집어넣으면 누구라도 실패함. 엑셀 공식이 길어지면 길어질수록 쪼개넣어야함. 쪼개서 넣어 확인하고 그 다음 또 쪼개서 넣기
* 여기서는 Age 컬럼으로 쪼개서 넣음
* 제일 많이 하는 실수가 IF 문에 3개를 넣어야하는데 4개를 넣거나 2개를 넣음
* 이걸 방지하기 위해 왼쪽부터 순서대로 공식을 쓰는게 아니라 괄호까지 아예 완성 즉, =IF(?, ?, ?) 미리 이렇게 만들어놓고 물음표를 바꾸는 형태로 공식 작성
* 먼저 =IF(E2 = "female, 1, 0) 넣어서 잘 돌아가는지 확인
* 그 다음 female이 참일 경우 들어가는 1에 다시 공식 삽입 =IF(E2 = "female", IF(?, ?, ?), 0)
* 즉 =IF(E2 = "female", IF(C2 = 3, 0, 1), 0)
* C~L열 모두 삭제하고 케글에 제출. Score = 75% 결과가 더 안좋아짐
* 결과가 더 안좋아진 이유: 3등급에 탄 여성의 생존률(50%)이 상대적으로(1등급에 탄 여성보다)낮았지만 절대적으로는 높음. 반대로 1등급에 탄 남성의 생존률은(36%) 다른 등급에 탄 남성보다 높았지만 절대적으로는 낮음.
