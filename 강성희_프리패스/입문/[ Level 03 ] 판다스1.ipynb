```python
# 안비어있는 걸 가져오고 싶을 땐

data[data['몸무게(prev)'].notnull()]
```

```python
# 조건이 여러 개 일 때
# 이렇게 하면 에러남
# 파이썬 and와 판다스이 서로 호환이 안됨

data['몸무게'] > 75.5 and data['담당'] == '김지수'
```

```python
# &(and, shift + 7) 이것도 실행안됨
# 가운데 &가 먼저 실행되므로 &왼쪽과 오른쪽이 먼저 실행되도록 ()로 감싸줘야함

(data['몸무게'] > 75.5) & (data['담당'] == '김지수')
```

```python
data[(data['몸무게'] > 75.5) & (data['담당'] == '김지수')]
```

```python
# 근데 이러면 가로로 길어짐
# 그래서 조건이 여러개인 경우 조건마다 변수로 빼주는게 좋음

high = (data['몸무게'] > 75.5)
kim = (data['담당'] == '김지수')

data[high & kim]
```

```python
# | (or, shift + 대괄호 닫는 버튼 오른쪽 키)

data[high | kim]
```

```python
# 담당선생님이 김지수인 경우 중에서 몸무게 75이상인경우

# 먼저 권장하지 않는 방식, 대괄호를 두번 열고 닫으므로 두번 실행, 시간 오래걸림

data[data['담당'] == '김지수']['몸무게']
```

```python
# 좋은 방식

data.loc[data['담당'] == '김지수', '몸무게']
```

```python
# loc에 ,(콤마)가 들어가면 row와 column인데 위의 경우 색인 후 column까지..
# 기능 여러가지이므로 헷갈리지 않게!
```

```python
data.loc[data['담당'] == '김지수', ['이름', '측정일', '몸무게']]
```

```python
columns = ['이름', '측정일', '몸무게']


data.loc[data['담당'] == '김지수', columns]
```

pd.pivot_table(data, index = '담당', values = '몸무게')

# 담당 선생님별 측정한 몸무게가 나옴

```python
pd.pivot_table(data, index = ['지점', '담당'], values = '몸무게')
```

```python
# 디폴트는 mean, 근데 이걸 누적치로 바꿀 수도 있음

pd.pivot_table(data, index = ['지점', '담당'], values = '몸무게', aggfunc = 'sum')
```

```python
# mean과 sum 동시에 넣을 수도 있음

pd.pivot_table(data, index = ['지점', '담당'], values = '몸무게', aggfunc = ['mean', 'sum'])
```

## 컬럼 추가 & 수정

```python
# 추가하는 방법과 수정하는 방법 똑같음
```

```python
# 추가하는 값이 모두 똑같은 경우

data['지역'] = '서울'
data
```

```python
# 서울을 부산으로 바꾸는 경우

data['지역'] = '부산'
data
```

```python
# 모두 같지 않은 경우, 갯수만 맞으면 알아서 들어감

data['결제'] = [True, True, True, True, False, True, True, False, True,]

data
```

```python
# 갯수가 조금이라도 틀리면 안들어감

data['결제'] = [True, True, True, True, False, True, True, False]
```

```python
# 매번 리스트 값을 수기로 넣기가 쉽지 않음
# 실제론 추가하는 방법
# 몸무게가 g단위로 필요할 때

data['몸무게'] * 1000
```

```python
data['몸무게(g)'] = data['몸무게'] * 1000
data
```

```python
# 조건으로 넣는 경우

data['과체중'] = data['몸무게'] > 75.8
data
```

```python
# 이 방식의 단점: True, False 외엔 넣기가 힘들다
```

```python
# True, False 외에 넣는 방법

data.loc[data['몸무게'] > 75.8, '상태'] = '과체중'

data.loc[data['몸무게'] <= 75.8, '상태'] = '정상'

data
```
